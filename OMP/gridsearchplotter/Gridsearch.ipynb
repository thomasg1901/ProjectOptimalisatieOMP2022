```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns


def heatmap_file(file, alpha_min, alpha_max, T_min, T_max):
    alpha = []
    T = []
    cost = []

    with open(file) as file_in:
        lines = []
        for line in file_in:
            lines.append(line)
            alpha.append(float("{0:.2f}".format(float(line.split("alpha = ")[1].split(" and T")[0]))))
            T.append(float(line.split("T = ")[1].split(" : ")[0]))
            cost.append(float(line.split(" : ")[1]))

    df = pd.DataFrame(list(zip(alpha, T, cost)),
                   columns =['alpha', 'T', 'cost'])
    
    df = df[df["alpha"] >= alpha_min]
    df = df[df["alpha"] <= alpha_max]
    
    df = df[df["T"] >= T_min]
    df = df[df["T"] <= T_max]

    df = df.pivot('alpha', 'T', 'cost')
    sns.set(rc = {'figure.figsize':(10,10)})
    ax = sns.heatmap(df)

```

```python
heatmap_file("gridsearch_A-400-90.txt", 0, 1, 0, 20000)

#Best result: 0.7 op 6300
```
