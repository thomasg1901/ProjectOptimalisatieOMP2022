```python
import matplotlib.pyplot as plt
import numpy as np
def plotter(time,cost):
    datapoints= np.array(cost,dtype=float)
    xpoints = np.array(time,dtype=float)

    plt.plot(xpoints, datapoints)
    plt.show()
```

```python
file = "../../A-100-30"

f = open(file+".txt", "r")
data = f.read().split(";")
data = data[:-1]

f = open(file+"_times.txt", "r")
x = f.read().split(";")
x = x[:-1]

plotter(x,data)
```
